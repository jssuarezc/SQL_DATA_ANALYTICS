```python
import os
import sqlite3
import pandas as pd

db_name="exercise42.db"
conn=sqlite3.connect(db_name)
cursor=conn.cursor()

cursor.execute("""CREATE TABLE IF NOT EXISTS phonelog(callerid INT, recipientid INT, datecalled DATETIME)""")
conn.commit()

pl_data=[
    (1, 2, '2019-01-01 09:00:00.000'),
       (1, 3, '2019-01-01 17:00:00.000'),
       (1, 4, '2019-01-01 23:00:00.000'),
       (2, 5, '2019-07-05 09:00:00.000'),
       (2, 3, '2019-07-05 17:00:00.000'),
       (2, 3, '2019-07-05 17:20:00.000'),
       (2, 5, '2019-07-05 23:00:00.000'),
       (2, 3, '2019-08-01 09:00:00.000'),
       (2, 3, '2019-08-01 17:00:00.000'),
       (2, 5, '2019-08-01 19:30:00.000'),
       (2, 4, '2019-08-02 09:00:00.000'),
       (2, 5, '2019-08-02 10:00:00.000'),
       (2, 5, '2019-08-02 10:45:00.000'),
       (2, 4, '2019-08-02 11:00:00.000')
]
cursor.executemany("INSERT INTO phonelog VALUES(?,?,?)",pl_data)
conn.commit()

querytoexec="SELECT * FROM phonelog"
df=pd.read_sql(querytoexec,conn)
print(df)
```

```python
querytoexec="WITH calls AS (SELECT callerid, DATE(datecalled) AS called_date, MIN(datecalled) AS first_call, MAX(datecalled) AS last_call FROM phonelog GROUP BY callerid, DATE(datecalled)) SELECT c.*, p1.recipientid FROM calls AS c INNER JOIN phonelog AS p1 ON c.callerid= p1.callerid AND c.first_call=p1.datecalled INNER JOIN phonelog AS p2 ON c.callerid= p2.callerid AND c.last_call=p2.datecalled WHERE p1.recipientid=p2.recipientid"
df=pd.read_sql(querytoexec,conn)
print(df)
conn.close()
```
