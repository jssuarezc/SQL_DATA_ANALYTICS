```python
import os
import sqlite3
import pandas as pd

db_name="exercise47.db"
conn=sqlite3.connect(db_name)
cursor=conn.cursor()

cursor.execute("""CREATE TABLE IF NOT EXISTS purchase_history(userid INT, productid INT, purchasedate DATE)""")
conn.commit()


p_data=[
    (1,1,'23-01-2012'),
    (1,2,'23-01-2012'),
    (1,3,'25-01-2012'),
    (2,1,'23-01-2012'),
    (2,2,'23-01-2012'),
    (2,2,'25-01-2012'),
    (2,4,'25-01-2012'),
    (3,4,'23-01-2012'),
    (3,1,'23-01-2012'),
    (4,1,'23-01-2012'),
    (4,2,'25-01-2012')
]
cursor.executemany("INSERT INTO purchase_history VALUES(?,?,?)",p_data)
conn.commit()

querytoexec="SELECT * FROM purchase_history"
df=pd.read_sql(querytoexec,conn)
print(df)
```

```python
querytoexec="SELECT userid FROM purchase_history GROUP BY userid HAVING COUNT(DISTINCT purchasedate)>1 AND COUNT(productid)=COUNT(DISTINCT productid)"
df=pd.read_sql(querytoexec,conn)
print(df)
conn.close()
```
