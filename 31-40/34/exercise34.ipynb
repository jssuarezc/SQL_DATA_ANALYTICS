```python
import sqlite3
import pandas as pd

db_name="exercise24.db"
conn=sqlite3.connect(db_name)
cursor=conn.cursor()

cursor.execute("""CREATE TABLE IF NOT EXISTS employee(emp_id INT, company VARCHAR(10), salary INT)""")
conn.commit()

emp_data=[
    (1,'A',2341),
    (2,'A',341),
    (3,'A',15),
    (4,'A',15314),
    (5,'A',451),
    (6,'A',513),
    (7,'B',15),
    (8,'B',13),
    (9,'B',1154),
    (10,'B',1345),
    (11,'B',1221),
    (12,'B',234),
    (13,'C',2345),
    (14,'C',2645),
    (15,'C',2645),
    (16,'C',2652),
    (17,'C',65)
]
cursor.executemany("INSERT INTO employee VALUES(?,?,?)",emp_data)

querytoexec="SELECT * FROM employee"
df=pd.read_sql(querytoexec,conn)
print(df)
```

```python
querytoexec="SELECT company, AVG(salary) FROM (SELECT *, row_number() OVER(PARTITION BY company ORDER BY salary) AS rn, COUNT(1) OVER (PARTITION BY company) AS total_cnt FROM employee) AS a WHERE rn BETWEEN total_cnt * 1.0 / 2 AND total_cnt * 1.0/ 2 + 1 GROUP BY company"
df=pd.read_sql(querytoexec,conn)
print(df)
```
