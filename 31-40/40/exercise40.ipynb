```python
import os
import sqlite3
import pandas as pd

db_name="exercise40.db"
conn=sqlite3.connect(db_name)
cursor=conn.cursor()

cursor.execute("""CREATE TABLE IF NOT EXISTS brands(category VARCHAR(20), brand_name VARCHAR(20))""")
conn.commit()

b_values=[
    ('chocolates','5-star'),
    (None,'dairy milk'),
    (None,'perk'),
    (None,'eclair'),
    ('Biscuits','britannia'),
    (None,'good day'),
    (None,'boost')
]
cursor.executemany("INSERT INTO brands VALUES(?,?)",b_values)

querytoexec="SELECT * FROM brands"
df=pd.read_sql(querytoexec,conn)
print(df)
```

```python
querytoexec="WITH cte1 AS (SELECT *, ROW_NUMBER() OVER (ORDER BY (SELECT NULL)) AS rn FROM brands), cte2 AS (SELECT *, LEAD(rn, 1) OVER (ORDER BY rn) AS next_rn FROM cte1 WHERE category IS NOT NULL) SELECT cte2.category, cte1.brand_name FROM cte1 INNER JOIN cte2 ON cte1.rn >= cte2.rn AND (cte1.rn <= cte2.next_rn - 1 OR cte2.next_rn IS NULL)"
df=pd.read_sql(querytoexec,conn)
print(df)
conn.close()
```
